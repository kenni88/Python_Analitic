```python
import pandas as pd
df = pd.read_csv('laptop_price.csv', sep = ',', encoding = 'latin-1')
df.tail(5)
```

```python
df["Cpu_Company"]=df["Cpu"].apply(lambda x: x.split()[0])
```

```python
df["Cpu_Company"].value_counts()
```

```python
def convert(x):
    x = x.split()[0]
    if x.endswith('GB'):
        res = x[:-2]
        # res = x.replace('GB', '')
    elif x.endswith('TB'):
        res = float(x[:-2])*1024
    return int(res)    

df['Memory_Amount'] = df['Memory'].apply(convert)
```

```python
df['Memory_Amount'].value_counts()
```

```python
def convert2(x):
    x = x.split()
    x = x[1:]
    return ' '.join(x)

df['Memory_Type'] = df['Memory'].apply(convert2)
```

```python
df['Memory_Type'].value_counts()
```

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace=True)
```

```python
df.info()
```

```python
df['SSD'] = 0
df.loc[df['Memory_Type'] == 'SSD', 'SSD'] = 1
```

```python
df.head()
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float64')
```
